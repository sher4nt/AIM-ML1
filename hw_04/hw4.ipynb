```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import OneHotEncoder
```

```python
num_features = data.select_dtypes(np.number).drop(columns=['target', 'sk_id_curr']).columns
num_features
```

```python
cat_features = data.select_dtypes('object').columns
cat_features, len(cat_features)
```

```python
num_transformer = Pipeline([
    ('imp_mean', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
])
cat_transformer = Pipeline(steps=[
    ('imp_most_freq', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[
        ('num', num_transformer, list(num_features)),
        ('cat', cat_transformer, list(cat_features)),
    ]
)
```

```python
pipeline = Pipeline([
    ('ct', ct),
    ('model',  LogisticRegression(random_state=0, n_jobs=-1, max_iter=2000)),
])
```

```python
pipeline.fit(data_dict['tr'], data_dict['tr'].target)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(data_dict['tst'].target, pipeline.predict_proba(data_dict['tst'])[:, 1])
```

Адекватные значения данного признака - не очень большие по модулю отрицательные значения, так как данные измеряются в днях. Как видно, максимальное значение равно 365243, что во-первых слишком много, а во-вторых положительное, и эти данные не поддаются трактовке.

```python
data_dict['tr'].DAYS_EMPLOYED.describe()
```

Поэтому заполним все положительные значения в этой колонке пропусками, с которыми будут работать другие трансформеры данных.

```python
# ваш код
from sklearn.base import BaseEstimator,TransformerMixin

class AnomalyEmployment(BaseEstimator, TransformerMixin):
    ''' 
    Данный трансформер исправляет значения в колонке 'DAYS_EMPLOYED'
    Любые положительные значения заменяются на np.nan
    '''
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        X.loc[X['DAYS_EMPLOYED'] > 0, 'DAYS_EMPLOYED'] = np.nan
        return X
```

Также хочется превратить признак `DAYS_BIRTH` (т.е. возраст) в категориальный, разбив его на бины и сделав One-Hot-Encoding, поэтому вычеркнем этот признак из списка числовых. Также на основании скоров клиента из внешних источников сделаем полиномиальные признаки.

```python
num_features = num_features.drop('DAYS_BIRTH')
poly_features = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
```

```python
class AddNanInfo(BaseEstimator, TransformerMixin):
    ''' 
    Данный трансформер создаёт для каждого признака колонку 
    со значением 1, если имеется пропуск, и 0 иначе
    '''
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        temp_df = pd.DataFrame([])
        for column in X.columns:
            nan_col = pd.isna(X[column])
            temp_df = pd.concat(
                (temp_df, nan_col.astype(int).rename(column + "_isna")), axis=1
            )
        return temp_df
```

```python
nan_info_transformer = Pipeline(
    [
        ("anomaly_emp", AnomalyEmployment()),
        ("add_nan_info", AddNanInfo()),
    ]
)

poly_transformer = Pipeline(
    [
        ("imp_mean", SimpleImputer(strategy="mean")),
        ("poly", PolynomialFeatures(degree=3)),
        ("scaler", StandardScaler()),
    ]
)

num_transformer = Pipeline(
    [
        ("anomaly_emp", AnomalyEmployment()),
        ("imp_mean", SimpleImputer(strategy="mean")),
        ("scaler", StandardScaler()),
    ]
)

cat_transformer = Pipeline(
    steps=[
        ("imp_most_freq", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(handle_unknown="ignore")),
    ]
)

age_transformer = Pipeline(
    [
        (
            "age_bins",
            KBinsDiscretizer(
                n_bins=18,
                strategy="uniform",
                subsample=None,
                encode="onehot-dense",
            ),
        )
    ]
)

ct = ColumnTransformer(
    transformers=[
        ("nan_info", nan_info_transformer, list(data_dict["tr"].columns)),
        ("poly", poly_transformer, poly_features),
        ("age", age_transformer, ["DAYS_BIRTH"]),
        ("num", num_transformer, list(num_features)),
        ("cat", cat_transformer, list(cat_features)),
    ]
)

pipeline = Pipeline(
    [
        ("ct", ct),
        (
            "model",
            LogisticRegression(C=0.1, random_state=0, n_jobs=-1, max_iter=2000),
        ),
    ]
)

pipeline.fit(data_dict["tr"], data_dict["tr"].target)
```

Выбейте на тесте больше `0.752` [roc_auc_score](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.roc_auc_score.html).
